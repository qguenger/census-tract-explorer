```python
import json

import geopandas as gpd
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import plotly.graph_objects as go
import plotly.express as px    # â for the default qualitative palette

# âââ 1. LOAD & PREP âââââââââââââââââââââââââââââââââââââââââââââââââââââââââââââ
tracts_gdf = gpd.read_file("census_tracts.geojson")
tracts_gdf['GEOID'] = tracts_gdf['GEOID'].astype(str)

data = pd.read_excel("clean_middleneighborhood_data.xlsx")
data['GEOID'] = data['Tract GEOID'].astype(str)
data['Year']  = pd.to_numeric(data['Year'], errors='coerce')

metrics = [
    "Median home value",
    "Median household income",
    "Median monthly contract rent",
    "Vacant housing units"
]

all_city = (
    data
    .groupby("Year", as_index=False)[metrics]
    .mean()
    .assign(GEOID="All Pittsburgh")
)

combined = pd.concat(
    [ data[["GEOID","Year"] + metrics], all_city ],
    ignore_index=True
)

geojson    = json.loads(tracts_gdf.to_json())
geoid_list = tracts_gdf['GEOID'].tolist()
n          = len(geoid_list)

# âââ 2. INITIAL MAP FIGURE âââââââââââââââââââââââââââââââââââââââââââââââââââââââ
INITIAL_LINE_COLOR = ["darkgray"] * n
INITIAL_LINE_WIDTH = [0.5]  * n

base_map = go.Figure(
    go.Choroplethmapbox(
        geojson=geojson,
        locations=geoid_list,
        z=[0]*n,
        featureidkey="properties.GEOID",
        marker_opacity=0.5,
        marker_line_color=INITIAL_LINE_COLOR,
        marker_line_width=INITIAL_LINE_WIDTH,
        customdata=geoid_list,
        showscale=False
    )
)
base_map.update_layout(
    mapbox_style="open-street-map",
    mapbox_zoom=11.5,
    mapbox_center={"lat":40.4406,"lon":-79.9959},
    margin={"r":0,"t":30,"l":0,"b":0},
    title="Click a Tract to (De)Select",
    clickmode='event+select'
)

# âââ 3. DASH APP ââââââââââââââââââââââââââââââââââââââââââââââââââââââââââââââââ
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H1("Pittsburgh Tract Explorer"),

    dcc.Store(id="selected-tracts", data=[]),

    html.Div(
        dcc.Graph(id="map", figure=base_map),
        style={'width':'60%','margin':'auto'}
    ),

    html.Div([
        dcc.Graph(id='ts-home-value'),
        dcc.Graph(id='ts-household-income'),
        dcc.Graph(id='ts-contract-rent'),
        dcc.Graph(id='ts-vacant-units'),
    ], style={
        'display':'grid',
        'gridTemplateColumns':'1fr 1fr',
        'gridTemplateRows':'auto auto',
        'gap':'20px',
        'width':'80%',
        'margin':'40px auto'
    })
])

# âââ 4. CALLBACK: TOGGLE & RESTYLE MAP ââââââââââââââââââââââââââââââââââââââââââ
@app.callback(
    Output("map", "figure"),
    Output("selected-tracts", "data"),
    Input("map", "clickData"),
    State("selected-tracts", "data")
)
def toggle_tract(clickData, selected):
    selected = selected or []
    if clickData and clickData.get("points"):
        geoid = clickData["points"][0]["location"]
        if geoid in selected:
            selected.remove(geoid)
        else:
            selected.append(geoid)

    # pick from the Plotly qualitative palette
    palette = px.colors.qualitative.Plotly

    # assign each selected GEOID its own color; others stay white
    new_colors = [
        palette[selected.index(g) % len(palette)] if g in selected else "white"
        for g in geoid_list
    ]
    new_widths = [
        3 if g in selected else 0.5
        for g in geoid_list
    ]

    fig = go.Figure(
        go.Choroplethmapbox(
            geojson=geojson,
            locations=geoid_list,
            z=[0]*n,
            featureidkey="properties.GEOID",
            marker_opacity=0.5,
            marker_line_color=new_colors,
            marker_line_width=new_widths,
            customdata=geoid_list,
            showscale=False
        )
    )
    fig.update_layout(
        mapbox_style="open-street-map",
        mapbox_zoom=11,
        mapbox_center={"lat":40.4406,"lon":-79.9959},
        margin={"r":0,"t":30,"l":0,"b":0},
        title="Click a Tract to (De)Select",
        clickmode='event+select'
    )
    return fig, selected

# âââ 5. CALLBACK: UPDATE 4 TIME-SERIES WITH MATCHING COLORS âââââââââââââââââââ
@app.callback(
    Output('ts-home-value', 'figure'),
    Output('ts-household-income', 'figure'),
    Output('ts-contract-rent', 'figure'),
    Output('ts-vacant-units', 'figure'),
    Input("selected-tracts", "data")
)
def update_timeseries(selected):
    geos = selected or ["All Pittsburgh"]
    palette = px.colors.qualitative.Plotly

    figs = []
    for m in metrics:
        f = go.Figure()
        for idx, g in enumerate(geos):
            df = combined[combined.GEOID == g]
            color = palette[idx % len(palette)]
            f.add_trace(go.Scatter(
                x=df["Year"],
                y=df[m],
                mode="lines+markers",
                name=g,
                line=dict(color=color),
                marker=dict(color=color)
            ))
        f.update_layout(
            title=m,
            xaxis_title="Year",
            yaxis_title=m
        )
        figs.append(f)

    return figs  # returns the four figures in order

if __name__ == "__main__":
    app.run_server(debug=True)
```
